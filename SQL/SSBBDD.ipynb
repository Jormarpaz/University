Volvemos a escribir código en el archivo, pero esta vez para crear la tabla, la cual será interna al no especificarla con "EXTERNAL", y es una decisión que facilita la gestión de los datos directamente desde Hive,
sin preocuparse de archivos fuera de su control. Útil sobre todo porque los datos se generan para el uso dentro del propio sistema. Es decir, la elección de una tabla interna permite a Hive procesar los archivos de una manera
eficiente. De haber elegido una tabla externa, tendría que haber usado la clausa EXTERNAL y haber especificado una ubicación, además, con ese método no se borrarían los datos al eliminar una tabla, lo cual implicaría un mayor
cuidado en la limpieza de información.

```python
%%writefile ssbbdd.hql
use convocatoriasdb; 

CREATE TABLE IF NOT EXISTS convocatorias 
( 
    Numero_Procedimiento STRING,
    Numero_SAF INT,
    Descripcion_SAF STRING,
    Numero_UOC INT,
    Descripcion_UOC STRING,
    Tipo_Procedimiento STRING,
    Modalidad STRING,
    Apartado_Directa STRING,
    Ejercicio INT,
    Fecha_Publicacion STRING,
    Fecha_Apertura STRING,
    Etapa STRING,
    Alcance STRING,
    Nombre_Procedimiento STRING,
    Objeto_Procedimiento STRING,
    Monto_Estimado FLOAT
) 
COMMENT 'Tabla de Convocatorias' 
ROW FORMAT DELIMITED 
FIELDS TERMINATED BY ','  
```

Volvemos a ejecutar el archivo para que se cree la tabla y mostramos las tablas existentes para asegurarnos

## Apartado 3:

Crear las consultas necesarias para responder las siguientes cuestiones:

### Devuelve los registros cuyo tipo de procedimiento de contratación sea contratación directa

```python
! beeline -u "jdbc:hive2://localhost:10000/convocatoriasdb" -e \
"select Numero_Procedimiento, Numero_SAF,Tipo_Procedimiento from convocatorias where Tipo_Procedimiento = 'Contratación Directa';"
```

### Devuelve los registros cuyo tipo de procedimiento sea contratación directa y cuyo ejercicio fiscal sea anterior a 2020

```python
! beeline -u "jdbc:hive2://localhost:10000/convocatoriasdb" -e \
"select Numero_Procedimiento, Numero_SAF,Tipo_Procedimiento, Ejercicio from convocatorias where Tipo_Procedimiento = 'Contratación Directa' and Ejercicio < 2020;"
```

### Cuantos tipos de SAF hay para los registros de las convocatorias?

```python
! beeline -u "jdbc:hive2://localhost:10000/convocatoriasdb" -e \
"select count(distinct Descripcion_SAF) as Tipos_SAF from convocatorias;"
```

### Cual es el maximo presupuesto estimado de estas convocatorias

```python
! beeline -u "jdbc:hive2://localhost:10000/convocatoriasdb" -e \
"select max(Monto_Estimado) as Max_Presupuesto from convocatorias;"
```

### Cual es el moton estimado que se emplea en las convocatorias agrupado por alcance

```python
! beeline -u "jdbc:hive2://localhost:10000/convocatoriasdb" -e \
"select Alcance, avg(Monto_Estimado) as Promedio_Monto from convocatorias group by Alcance;"
```

## Apartado 4:

Desarrolla el codigo para Mapeduce en python que implemente la primera consulta "Devuelve los registros cuyo tipo de procedimiento de contratación sea contratación directa"

```python
%%writefile mapper.py
#!/usr/bin/env python
#-*- coding: utf-8 -*-
import sys

for line in sys.stdin:
    line = line.strip()
    columns = line.split(",")
    
    if len(columns) > 5:
        tipo_procedimiento = columns[5].strip()
        if tipo_procedimiento == "Contratación Directa":
            print(line)
```

```python
%%writefile reducer.py
#!/usr/bin/env python
#-*- coding: utf-8 -*-
import sys

for line in sys.stdin:
    print(line.strip())
```

### Consultas

La primera consulta mostrará una lista de titulos cuyo rating es superior a 8.

```python
!beeline -u "jdbc:hive2://localhost:10000/pelisdb" -e \
"select ts.Title, ts.Year, ts.Age, ts.Type, cast(regexp_replace(imdb_movies.rating, '/10', '') as float) as imdb_rating, cast(regexp_replace(ts.imdb_rating, '/10', '') as float) as tv_shows_rating from tv_shows ts join imdb_movies on ts.Title = imdb_movies.Title where cast(regexp_replace(imdb_movies.rating, '/10', '') as float) > 8 and cast(regexp_replace(ts.imdb_rating, '/10', '') as float) > 8;"
```

Selección de títulos disponibles en Netlix y Prime_Video que pertenezcan al año 2020 o más reciente.

```python
! beeline -u "jdbc:hive2://localhost:10000/pelisdb" -e \
"select ts.Title, ts.Year, ts.Age, ts.Type, imdb_movies.year as imdb_year from tv_shows ts join imdb_movies on ts.Title = imdb_movies.title where ts.Netflix = 1 and ts.Prime_Video = 1 and ts.Year >= 2020;"
```
