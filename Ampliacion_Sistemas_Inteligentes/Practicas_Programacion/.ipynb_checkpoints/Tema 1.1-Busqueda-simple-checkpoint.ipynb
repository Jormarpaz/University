# <font color="#004D7F" size=6> 1. Introducci贸n a la B煤squeda Local</font>

La b煤squeda local, tambi茅n conocida como algoritmos de mejora iterativa, es un tipo de **metaheur铆stica** que opera iterativamente, movi茅ndose de una soluci贸n candidata a una soluci贸n vecina en el espacio de b煤squeda. Estos m茅todos son cruciales para resolver problemas de optimizaci贸n combinatoria que son NP-duros.

Los algoritmos presentados a continuaci贸n pertenecen al grupo de **b煤squeda local** o **mejora iterativa**.

```python
import random
import math
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
import pandas as pd # type: ignore

# Definici贸n de una funci贸n de prueba con 贸ptimos locales (Landscape)
# La funci贸n debe ser suficientemente compleja para demostrar las debilidades de Hill-Climbing.
def funcion_objetivo(x):
    """
    Funci贸n de adecuaci贸n (fitness) a MAXIMIZAR.
    Ejemplo: Funci贸n con varios picos (un seno amortiguado, por ejemplo).
    """
    # Usamos una funci贸n que simula un "relieve" irregular [1].
    return np.sin(x) * np.exp(-0.01 * x) + (x * 0.1)

# Definici贸n del espacio de b煤squeda (vecindad)
def generar_vecino(x, paso=1.0):
    """Genera un vecino moviendo la soluci贸n actual ligeramente."""
    if random.random() < 0.5:
        return x + paso * random.uniform(0.1, 2.0)
    else:
        return x - paso * random.uniform(0.1, 2.0)

# Rango de b煤squeda para la visualizaci贸n
X_MIN = 0
X_MAX = 100
```

# <font color="#004D7F" size=6> 2. Algoritmo de Escalada (Hill-Climbing)</font>

Este algoritmo es la forma m谩s simple de b煤squeda local y es un m茅todo **voraz (*greedy*)**. En cada paso, toma decisiones irrevocables [26], avanzando siempre hacia el vecino que ofrece la **mejor mejora**.

**Limitaci贸n Te贸rica Clave**: Su principal limitaci贸n es que se detiene al alcanzar un **贸ptimo local** (un punto donde todos los vecinos inmediatos son peores), ya que carece de memoria y no considera caminos alternativos. No garantiza la completitud ni la admisibilidad.
S

```python
def hill_climbing_mod(inicio, max_iteraciones=1000):
    """
    Implementaci贸n del algoritmo de escalada que registra la trayectoria.
    """
    actual = inicio
    valor_actual = funcion_objetivo(actual)
    
    x_history = [actual]
    f_history = [valor_actual]

    for i in range(max_iteraciones):
        mejor_vecino = None
        mejor_valor_vecino = valor_actual

        # Explorar vecinos
        for _ in range(5): # Reducimos la exploraci贸n a 5 vecinos por eficiencia
            vecino = generar_vecino(actual)
            valor_vecino = funcion_objetivo(vecino)

            if valor_vecino > mejor_valor_vecino:
                mejor_valor_vecino = valor_vecino
                mejor_vecino = vecino

        # Condici贸n de avance: solo si encontramos una mejora
        if mejor_valor_vecino > valor_actual:
            actual = mejor_vecino
            valor_actual = mejor_valor_vecino
        else:
            # ptimo local alcanzado
            break
            
        x_history.append(actual)
        f_history.append(valor_actual)

    return actual, valor_actual, x_history, f_history
```

# <font color="#004D7F" size=6> 3. Temple Simulado (Simulated Annealing)</font>

El Temple Simulado, o *Simulated Annealing* (SA)], resuelve el problema del 贸ptimo local en *Hill-Climbing*. SA se inspira en el proceso metal煤rgico de enfriamiento gradual y permite la aceptaci贸n de soluciones peores, con una **probabilidad que decrece** a medida que avanza la b煤squeda (disminuye la *temperatura* $T$).

**Mecanismo Clave (Teor铆a)

La clave es la **regla de aceptaci贸n de Metropolis**, que define la probabilidad $P$ de aceptar un movimiento a un estado peor ($\Delta E < 0$):

$$ P(\text{aceptar}) = e^{\Delta E / T} $$

Donde $\Delta E$ es la diferencia de valor entre el estado nuevo y el actual, y $T$ es la temperatura. Cuanto menor sea $T$ o mayor sea la p茅rdida ($\Delta E$ muy negativo), menor ser谩 la probabilidad de aceptaci贸n.

```python
def simulated_annealing_mod(inicio, max_iteraciones=5000, temp_inicial=100.0, factor_enfriamiento=0.995):
    """
    Implementaci贸n del Temple Simulado que registra la trayectoria.
    Ajustamos las iteraciones y el enfriamiento para una mejor demostraci贸n.
    """
    actual = inicio
    mejor_global = inicio
    T = temp_inicial

    x_history = [actual]
    f_history = [funcion_objetivo(actual)]
    
    for i in range(max_iteraciones):
        T *= factor_enfriamiento # Enfriamiento gradual [2]

        if T < 0.0001:
            break

        vecino = generar_vecino(actual)
        
        valor_actual = funcion_objetivo(actual)
        valor_vecino = funcion_objetivo(vecino)

        delta_E = valor_vecino - valor_actual

        # Regla de Metropolis
        if delta_E > 0 or math.exp(delta_E / T) > random.random():
            actual = vecino
            
        # Actualizar el mejor global
        if funcion_objetivo(actual) > funcion_objetivo(mejor_global):
            mejor_global = actual

        x_history.append(actual)
        f_history.append(funcion_objetivo(actual))
        
    return mejor_global, funcion_objetivo(mejor_global), x_history, f_history
```

# <font color="#004D7F" size=6>4. B煤squeda Tab煤 (Tabu Search)</font>

La **B煤squeda Tab煤 (BT)** es una metaheur铆stica que utiliza **memoria** para guiar la b煤squeda fuera de las regiones ya exploradas. A diferencia de *Hill-Climbing* y *Simulated Annealing*, BT mantiene una **Lista Tab煤** que almacena las soluciones (o movimientos) que se han visitado recientemente y que deben ser evitadas durante un cierto n煤mero de iteraciones. Esto previene el **ciclado** y obliga al algoritmo a explorar otras zonas del espacio de b煤squeda.

**Mecanismo Clave (Teor铆a)**

1.  **Lista Tab煤:** Guarda movimientos "prohibidos" por un n煤mero $L$ de iteraciones (tenencia tab煤).
2.  **Criterio de Aspiraci贸n (opcional pero com煤n):** Permite anular la restricci贸n tab煤 si el movimiento prohibido conduce a una soluci贸n **mejor que el mejor global** registrado hasta el momento.


```python
def tabu_search_mod(inicio, max_iteraciones=1000, tamano_lista_tabu=15):
    """
    Implementaci贸n de B煤squeda Tab煤 que registra la trayectoria.
    """
    actual = inicio
    mejor_global = inicio
    lista_tabu = []

    x_history = [actual]
    f_history = [funcion_objetivo(actual)]

    for i in range(max_iteraciones):
        
        mejor_vecino_ciclo = None
        mejor_valor_vecino_ciclo = -float('inf')
        
        vecinos = [generar_vecino(actual) for _ in range(20)] 

        for vecino in vecinos:
            valor_vecino = funcion_objetivo(vecino)

            es_tabu = round(vecino) in [round(x) for x in lista_tabu] # Comparaci贸n aproximada
            
            # Criterio de Aspiraci贸n: Permite el movimiento tab煤 si mejora el 贸ptimo global [3]
            criterio_aspiracion = (valor_vecino > funcion_objetivo(mejor_global))

            if (not es_tabu) or criterio_aspiracion:
                if valor_vecino > mejor_valor_vecino_ciclo:
                    mejor_valor_vecino_ciclo = valor_vecino
                    mejor_vecino_ciclo = vecino

        # Si se encontr贸 un movimiento v谩lido (no tab煤 o con aspiraci贸n)
        if mejor_vecino_ciclo is not None:
            actual = mejor_vecino_ciclo
            
            # Actualizar Lista Tab煤
            if len(lista_tabu) >= tamano_lista_tabu:
                lista_tabu.pop(0) 
            lista_tabu.append(actual) 
            
            # Actualizar el mejor global
            if funcion_objetivo(actual) > funcion_objetivo(mejor_global):
                mejor_global = actual
        
        else:
            # Si no hay movimientos, la b煤squeda se estanca.
            break

        x_history.append(actual)
        f_history.append(funcion_objetivo(actual))


    return mejor_global, funcion_objetivo(mejor_global), x_history, f_history
```
