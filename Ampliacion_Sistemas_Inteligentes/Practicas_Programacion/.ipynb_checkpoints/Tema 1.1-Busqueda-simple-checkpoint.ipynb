# <font color="#004D7F" size=6> 1. Introducción a la Búsqueda Local</font>

La búsqueda local, también conocida como algoritmos de mejora iterativa, es un tipo de **metaheurística** que opera iterativamente, moviéndose de una solución candidata a una solución vecina en el espacio de búsqueda. Estos métodos son cruciales para resolver problemas de optimización combinatoria que son NP-duros.

Los algoritmos presentados a continuación pertenecen al grupo de **búsqueda local** o **mejora iterativa**.

```python
import random
import math
import numpy as np # type: ignore
import matplotlib.pyplot as plt # type: ignore
import pandas as pd # type: ignore

# Definición de una función de prueba con óptimos locales (Landscape)
# La función debe ser suficientemente compleja para demostrar las debilidades de Hill-Climbing.
def funcion_objetivo(x):
    """
    Función de adecuación (fitness) a MAXIMIZAR.
    Ejemplo: Función con varios picos (un seno amortiguado, por ejemplo).
    """
    # Usamos una función que simula un "relieve" irregular [1].
    return np.sin(x) * np.exp(-0.01 * x) + (x * 0.1)

# Definición del espacio de búsqueda (vecindad)
def generar_vecino(x, paso=1.0):
    """Genera un vecino moviendo la solución actual ligeramente."""
    if random.random() < 0.5:
        return x + paso * random.uniform(0.1, 2.0)
    else:
        return x - paso * random.uniform(0.1, 2.0)

# Rango de búsqueda para la visualización
X_MIN = 0
X_MAX = 100
```

<a id="section2"></a>
# <font color="#004D7F" size=6> 2. Algoritmo de Escalada (Hill-Climbing)</font>

Este algoritmo es la forma más simple de búsqueda local y es un método **voraz (*greedy*)**. En cada paso, toma decisiones irrevocables [26], avanzando siempre hacia el vecino que ofrece la **mejor mejora**.

**Limitación Teórica Clave**: Su principal limitación es que se detiene al alcanzar un **óptimo local** (un punto donde todos los vecinos inmediatos son peores), ya que carece de memoria y no considera caminos alternativos. No garantiza la completitud ni la admisibilidad.
S

```python
def hill_climbing_mod(inicio, max_iteraciones=1000):
    """
    Implementación del algoritmo de escalada que registra la trayectoria.
    """
    actual = inicio
    valor_actual = funcion_objetivo(actual)
    
    x_history = [actual]
    f_history = [valor_actual]

    for i in range(max_iteraciones):
        mejor_vecino = None
        mejor_valor_vecino = valor_actual

        # Explorar vecinos
        for _ in range(5): # Reducimos la exploración a 5 vecinos por eficiencia
            vecino = generar_vecino(actual)
            valor_vecino = funcion_objetivo(vecino)

            if valor_vecino > mejor_valor_vecino:
                mejor_valor_vecino = valor_vecino
                mejor_vecino = vecino

        # Condición de avance: solo si encontramos una mejora
        if mejor_valor_vecino > valor_actual:
            actual = mejor_vecino
            valor_actual = mejor_valor_vecino
        else:
            # Óptimo local alcanzado
            break
            
        x_history.append(actual)
        f_history.append(valor_actual)

    return actual, valor_actual, x_history, f_history
```

<a id="section3"></a>
# <font color="#004D7F" size=6> 3. Temple Simulado (Simulated Annealing)</font>

El Temple Simulado, o *Simulated Annealing* (SA)], resuelve el problema del óptimo local en *Hill-Climbing*. SA se inspira en el proceso metalúrgico de enfriamiento gradual y permite la aceptación de soluciones peores, con una **probabilidad que decrece** a medida que avanza la búsqueda (disminuye la *temperatura* $T$).

**Mecanismo Clave (Teoría)

La clave es la **regla de aceptación de Metropolis**, que define la probabilidad $P$ de aceptar un movimiento a un estado peor ($\Delta E < 0$):

$$ P(\text{aceptar}) = e^{\Delta E / T} $$

Donde $\Delta E$ es la diferencia de valor entre el estado nuevo y el actual, y $T$ es la temperatura. Cuanto menor sea $T$ o mayor sea la pérdida ($\Delta E$ muy negativo), menor será la probabilidad de aceptación.

```python
def simulated_annealing_mod(inicio, max_iteraciones=5000, temp_inicial=100.0, factor_enfriamiento=0.995):
    """
    Implementación del Temple Simulado que registra la trayectoria.
    Ajustamos las iteraciones y el enfriamiento para una mejor demostración.
    """
    actual = inicio
    mejor_global = inicio
    T = temp_inicial

    x_history = [actual]
    f_history = [funcion_objetivo(actual)]
    
    for i in range(max_iteraciones):
        T *= factor_enfriamiento # Enfriamiento gradual [2]

        if T < 0.0001:
            break

        vecino = generar_vecino(actual)
        
        valor_actual = funcion_objetivo(actual)
        valor_vecino = funcion_objetivo(vecino)

        delta_E = valor_vecino - valor_actual

        # Regla de Metropolis
        if delta_E > 0 or math.exp(delta_E / T) > random.random():
            actual = vecino
            
        # Actualizar el mejor global
        if funcion_objetivo(actual) > funcion_objetivo(mejor_global):
            mejor_global = actual

        x_history.append(actual)
        f_history.append(funcion_objetivo(actual))
        
    return mejor_global, funcion_objetivo(mejor_global), x_history, f_history
```

# <font color="#004D7F" size=6>4. Búsqueda Tabú (Tabu Search)</font>

La **Búsqueda Tabú (BT)** es una metaheurística que utiliza **memoria** para guiar la búsqueda fuera de las regiones ya exploradas. A diferencia de *Hill-Climbing* y *Simulated Annealing*, BT mantiene una **Lista Tabú** que almacena las soluciones (o movimientos) que se han visitado recientemente y que deben ser evitadas durante un cierto número de iteraciones. Esto previene el **ciclado** y obliga al algoritmo a explorar otras zonas del espacio de búsqueda.

**Mecanismo Clave (Teoría)**

1.  **Lista Tabú:** Guarda movimientos "prohibidos" por un número $L$ de iteraciones (tenencia tabú).
2.  **Criterio de Aspiración (opcional pero común):** Permite anular la restricción tabú si el movimiento prohibido conduce a una solución **mejor que el mejor global** registrado hasta el momento.


```python
def tabu_search_mod(inicio, max_iteraciones=1000, tamano_lista_tabu=15):
    """
    Implementación de Búsqueda Tabú que registra la trayectoria.
    """
    actual = inicio
    mejor_global = inicio
    lista_tabu = []

    x_history = [actual]
    f_history = [funcion_objetivo(actual)]

    for i in range(max_iteraciones):
        
        mejor_vecino_ciclo = None
        mejor_valor_vecino_ciclo = -float('inf')
        
        vecinos = [generar_vecino(actual) for _ in range(20)] 

        for vecino in vecinos:
            valor_vecino = funcion_objetivo(vecino)

            es_tabu = round(vecino) in [round(x) for x in lista_tabu] # Comparación aproximada
            
            # Criterio de Aspiración: Permite el movimiento tabú si mejora el óptimo global [3]
            criterio_aspiracion = (valor_vecino > funcion_objetivo(mejor_global))

            if (not es_tabu) or criterio_aspiracion:
                if valor_vecino > mejor_valor_vecino_ciclo:
                    mejor_valor_vecino_ciclo = valor_vecino
                    mejor_vecino_ciclo = vecino

        # Si se encontró un movimiento válido (no tabú o con aspiración)
        if mejor_vecino_ciclo is not None:
            actual = mejor_vecino_ciclo
            
            # Actualizar Lista Tabú
            if len(lista_tabu) >= tamano_lista_tabu:
                lista_tabu.pop(0) 
            lista_tabu.append(actual) 
            
            # Actualizar el mejor global
            if funcion_objetivo(actual) > funcion_objetivo(mejor_global):
                mejor_global = actual
        
        else:
            # Si no hay movimientos, la búsqueda se estanca.
            break

        x_history.append(actual)
        f_history.append(funcion_objetivo(actual))


    return mejor_global, funcion_objetivo(mejor_global), x_history, f_history
```

# <font color="#004D7F" size=6> 5. Comparación y Visualización de la Eficiencia Algorítmica</font>

Para el ingeniero informático, es esencial evaluar la **calidad de la solución** y la **eficiencia computacional** de estas metaheurísticas. La visualización nos permite analizar cómo navega cada algoritmo por el *landscape* de la función objetivo.

**Ejecución de los Algoritmos (Punto de Partida Fijo)**: Estableceremos un punto de partida (semilla) y un rango de búsqueda fijos para una comparación justa.

```python
random.seed(42) # Fija la semilla para reproducibilidad
inicio = 15.0

# Ejecutar los algoritmos modificados
_, _, hc_x, hc_f = hill_climbing_mod(inicio=inicio, max_iteraciones=2000)
_, _, sa_x, sa_f = simulated_annealing_mod(inicio=inicio, max_iteraciones=2000)
_, _, ts_x, ts_f = tabu_search_mod(inicio=inicio, max_iteraciones=2000)

print(f"Valor final HC: f(x)={hc_f[-1]:.4f} después de {len(hc_f)} iteraciones")
print(f"Valor final SA: f(x)={sa_f[-1]:.4f} después de {len(sa_f)} iteraciones")
print(f"Valor final TS: f(x)={ts_f[-1]:.4f} después de {len(ts_f)} iteraciones")

resultados = {
    "Hill Climbing": {"x": hc_x, "f": hc_f, "color": "blue"},
    "Simulated Annealing": {"x": sa_x, "f": sa_f, "color": "red"},
    "Tabu Search": {"x": ts_x, "f": ts_f, "color": "green"}
}

```

# <font color="#004D7F" size=6> Gráfico A: Convergencia (Evolución del Valor de la Función Objetivo)</font>

Este gráfico muestra el valor de la función objetivo (`f(x)`) a lo largo de las iteraciones.

*   **Hill-Climbing** debe mostrar una curva **monotónica no decreciente** hasta que se estanca.
*   **Simulated Annealing** debe mostrar **caídas abruptas** (saltos a soluciones peores), especialmente al principio, demostrando su capacidad de exploración.
*   **Búsqueda Tabú** también mostrará exploración, forzada por su mecanismo de **memoria**.


```python

plt.figure(figsize=(12, 6))
plt.title("Gráfico de Convergencia: Evolución de f(x) por Iteración")
plt.xlabel("Iteraciones")
plt.ylabel("Valor de la Función Objetivo (f(x))")

for nombre, data in resultados.items():
    plt.plot(data['f'], label=nombre, color=data['color'], alpha=0.7)

plt.legend()
plt.grid(True)
plt.show()
```

# <font color="#004D7F" size=6>Gráfico B: Visualización del Landscape (Recorrido de la Búsqueda)</font>

Si el problema es unidimensional, podemos graficar la función objetivo completa (el *landscape*) y superponer el camino de los algoritmos. Esto ilustra cómo cada algoritmo navega por el relieve del problema.

*   Observe cómo Hill-Climbing se queda atrapado en el primer pico cerca del punto de inicio.
*   Observe cómo Simulated Annealing y Búsqueda Tabú logran "saltar" y explorar picos más prometedores.


```python
# 1. Graficar el landscape de la función objetivo
x_range = np.linspace(X_MIN, X_MAX, 500)
y_range = funcion_objetivo(x_range)

plt.figure(figsize=(14, 7))
plt.plot(x_range, y_range, label="Función Objetivo (Landscape)", color='gray', linestyle='--', alpha=0.6)

# 2. Superponer los caminos de los algoritmos
for nombre, data in resultados.items():
    # Usamos la historia de X (posición) y la historia de F (valor)
    plt.plot(data['x'], data['f'], marker='.', linestyle='-', markersize=5, 
             label=f"{nombre} (Camino)", color=data['color'])

plt.scatter(inicio, funcion_objetivo(inicio), marker='o', s=100, color='black', label="Inicio")

plt.title("Visualización del Recorrido de los Algoritmos de Búsqueda Local")
plt.xlabel("Posición (x)")
plt.ylabel("Valor de la Función Objetivo (f(x))")
plt.legend()
plt.grid(True, linestyle=':', alpha=0.5)
plt.xlim(X_MIN, X_MAX)
plt.show()
```
